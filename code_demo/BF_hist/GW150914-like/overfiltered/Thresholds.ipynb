## Takeaway: downsampling does not change the threshold

```python
import numpy as np
import matplotlib.pyplot as pl
```

## 220+210:220

```python
data = np.loadtxt("results/GW150914_overfiltered_220+210:220_ALL.dat")
[SNRtot_MF, evidence_inj, evidence_filt, nofilter] = data.T
thresh = np.median(evidence_filt - evidence_inj)+2*np.std(evidence_filt - evidence_inj)
```

```python
pl.plot(SNRtot_MF, evidence_filt - evidence_inj, 'o')
pl.axhline(thresh, c='r', ls='--')
# pl.ylim(-100, 10)
```
