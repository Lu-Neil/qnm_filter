```python
import numpy as np
import matplotlib.pyplot as pl
import qnm
```

## NNI

```python
data = np.loadtxt("results/221_NNI_EC_ALL.dat")
[SNRtot, SNR220, SNR221, evidence220_221, evidence220, evidence221, nofilter] = data.T
idx = np.flip(np.argsort(SNRtot))
[SNRtot, SNR220, SNR221, evidence220_221, evidence220, evidence221, nofilter] = data[idx].T
time = np.arange(0, 100)
```

```python
thresh = 4.042081552641102
injected = (100, 0.4)
omega220 = qnm.modes_cache(s=-2,l=2,m=2,n=0)(a=injected[1])[0]
omega221 = qnm.modes_cache(s=-2,l=2,m=2,n=1)(a=injected[1])[0]
```

```python
tau221 = -1/np.imag(omega221)
```

```python
pl.plot(time, evidence220_221-evidence220)
pl.axhline(thresh, c='orange', ls='--')
# pl.axvline(tau221, c='r', ls='--')
# pl.axvline(2*tau221, c='r', ls='--')
# pl.axvline(3*tau221, c='r', ls='--')
pl.xlim(0, 20)
```

```python
BF = evidence220_221-evidence220
time[abs(BF-thresh).argmin()]
```

```python
pl.plot(time, SNRtot)
pl.axvline(tau221, c='r', ls='--')
# pl.axvline(2*tau221, c='r', ls='--')
pl.axvline(3*tau221, c='r', ls='--')
```
