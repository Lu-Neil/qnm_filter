# Cleaning Modes With Rational Filters

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
import matplotlib.pyplot as pl
import numpy as np
import qnm_filter
from gwpy.timeseries import TimeSeries
import copy
```

```python
data = np.loadtxt('Systematics_ALL.txt')
```

```python
pl.plot(data[0], data[1])
```

```python
t_array = data[0]
saved_log_evidence = data[1]
```

```python
mid_index = int(1*len(t_array)/2)
end_index = int(3*len(t_array)/4)
linear, cov = np.polyfit(t_array[mid_index:end_index], saved_log_evidence[mid_index:end_index], 0, cov = True)
intercept_index = np.argmax(saved_log_evidence > linear[0] - 5*cov[0])
```

```python
pl.plot(t_array, saved_log_evidence, 'o', ms=1.5, label='w/ filter')
pl.axhline(linear[0], c = 'r', ls = '--', label = 'Linear')
pl.axhline(linear[0] - 5*cov[0], c = 'g', ls = '--', label = '5 sigma')
# pl.axvline(t_array[index], c = 'y')
pl.legend()
pl.ylabel('lg Evidence')
pl.xlabel('t (ms)')
```

```python
t_array[intercept_index]
```
