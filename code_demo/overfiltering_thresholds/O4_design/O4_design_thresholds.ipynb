```python
import numpy as np
import matplotlib.pyplot as pl
import qnm
```

```python
data = np.loadtxt("results/O4_design_overfiltered_220+221:220_ALL.dat")
[SNRtot_MF, evidence_inj, evidence_filt, nofilter] = data.T
thresh = np.median(evidence_filt - evidence_inj)+3*np.std(evidence_filt - evidence_inj)
pl.plot(SNRtot_MF, evidence_filt - evidence_inj, 'o')
pl.axhline(thresh, c='r', ls='--')
# pl.ylim(-100, 10)
print(thresh)
```
