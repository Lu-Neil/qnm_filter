```python
import numpy as np
import matplotlib.pyplot as pl
import qnm
```

```python
data = np.loadtxt("results/O4_design_overfiltered_220+221:220_ALL.dat")
[SNRtot_MF, evidence_inj, evidence_filt, nofilter] = data.T
BF = evidence_filt-evidence_inj
thresh = np.median(BF)+3*np.std(BF)
pl.plot(SNRtot_MF, BF, 'o')
pl.axhline(thresh, c='r', ls='--')
# pl.ylim(-100, 10)
print(thresh)
```

```python
from scipy import stats
stats.normaltest(BF)
```

```python
BF = evidence_filt-evidence_inj
thresh_arr = []
std_arr = []
idx_arr = range(len(BF))
for i in idx_arr:
    temp = BF[0:i+1]
    thresh = np.quantile(temp, 0.95)
    thresh_arr.append(thresh)
for i in idx_arr:
    std_arr.append(np.std(thresh_arr[0:i]))
    
thresh_arr = np.asarray(thresh_arr)
std_arr = np.asarray(std_arr)
pl.plot(idx_arr, thresh_arr)
pl.fill_between(idx_arr, thresh_arr+std_arr, thresh_arr-std_arr, color='tab:blue', alpha=0.2)
pl.xlabel("Simulation number")
pl.ylabel("BF threshold")
```
